```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
Xs = np.array([1, 3, 5, 7, 9])

def f(x):
    return 5*x**2 - 6*x + 9

print(f(Xs))
```

## Markers

```python
plt.plot(Xs, f(Xs), marker = 'H', color = 'red')
plt.show()
```

You can also use the shortcut string notation parameter to specify the marker.

This parameter is also called fmt, and is written with this syntax:

`marker|line|color`

```python
plt.plot(f(Xs), 'o--m')
plt.show()

# o marker, -- dotted line, m magenta color
```

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 20, mfc = 'r', mec = 'black')
plt.show()
```

```python

plt.plot(ypoints, 'H-b', ms = 10, mfc = 'hotpink')
plt.show()
```

### plotting multiple plots on the same grid

```python
def y1(x):
    return x**2-2*x+5

def y2 (x):
    return 3*x**2-5

x = np.arange(-3,3,0.1)

plt.plot(x,y1(x), x,  y2(x), '*:b')
plt.show()
```

```python
plt.plot(x,y1(x), '2:' , color = 'hotpink')
plt.plot(x, y2(x), '*--g')
plt.show()
```

## Plot Labels and grids

`xlabel()`,` ylabel()`, and `title()` functions; use `fontdict` parameter for styling. 

```python
x1 = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y1 = np.array([240, 250, 260, 265, 280, 290, 300, 310, 320, 330])

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title("Sports Watch Data", fontdict = font1)
plt.xlabel("Average Pulse", fontdict = font2)
plt.ylabel("Calorie Burnage", fontdict = font2)

plt.plot(x1, y1)

plt.grid(axis='x', color = 'green', linestyle = '--', linewidth = 0.5) # select desired grid parameters

plt.show()
```

## Subplots

The `subplot()` function takes three arguments that describes the layout of the figure.

The layout is organized in rows and columns, which are represented by the first and second argument.

The third argument represents the index of the current plot.

Ex: `plt.subplot(1, 2, 1)`
#the figure has 1 row, 2 columns, and this plot is the first plot.

```python
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])
font2 = {'family':'serif','color':'black','size':15}

plt.subplot(1, 2, 1)
plt.plot(x,y)
plt.title("SALES", fontdict=font2)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y)
plt.title("INCOME", fontdict=font2)

plt.suptitle("MY SHOP", color = 'red')
plt.show()
```

for more specific colors, writing the hex code also works. 

## Scatter charts

- color of dots: `cmap` colormaps, diff colors for diff datasets on the same grid
- size of dots: `s` big/ small dot based on another parameter
- opacity: `alpha`

```python
# load a sample dataset with 2 features and a label. created using draw.xyz

import pandas as pd

# df = pd.read_csv('Data/data.csv', sep=',', header= 0)
df = pd.read_csv('Data/data.csv', header= 0)
```

```python
X = df.iloc[:, :2]
y = df.iloc[:, -1]
print(X.shape, y.shape)
```

```python
print(type(X))
print(type(y))
```

```python
y.nunique() # labels are either 'a' or 'b'
```

```python
# Convert categorical labels to numeric -- necessary to assign colormap!
y_numeric, y_labels = pd.factorize(y)  

scatter = plt.scatter(X.iloc[:, 0], X.iloc[:, 1], c=y_numeric, cmap='viridis', edgecolor='k', alpha=0.8)
plt.show()
```

## Piecharts

- Legends
- colors
- shadow
- explode
- start argument/ angle

```python
# a simple chart 

fruit_sales = np.array([10,35,48, 24, 60])
fruits = ['apple', 'mango', 'litchi', 'papita', 'ananas']

plt.pie(fruit_sales, labels=fruits)
plt.show()
```

```python
# explode, shadow and legend with title

fruit_sales = np.array([10,35,48, 24, 60])
fruits = ['apple', 'mango', 'litchi', 'papita', 'ananas']
explode_coeff = [0, 0.3 ,0 ,0.4,0]
shadow_bool = [True, False, False, True, False]

plt.pie(fruit_sales, labels=fruits, explode=explode_coeff, shadow = True)
plt.legend(title = "Fruit sales Jun'25", loc = 'lower left') 
plt.show()
```

```python
# custom color maps and starting angle
import matplotlib.cm as cm

# Example data
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10] 

# Choose a colormap (e.g., 'viridis', 'coolwarm', 'tab10', etc.)
cmap = cm.get_cmap('PRGn') 

# Generate colors from the colormap
colors = [cmap(i / len(sizes)) for i in range(len(sizes))]

# Create pie chart with custom colors
plt.figure(figsize=(6,6))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title("Pie Chart with Custom Colormap")
plt.show()
```
